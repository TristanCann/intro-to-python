## Loops

Now that we've seen the basics of working with a Jupyter notebook and the most common Python data types, it's time to begin putting this all together to start programming. Many computer programming tasks involve either repeatedly carrying out the same operation or performing different operations based on the current conditions. This section will introduce common methods for creating loops.

### `for` loops

`for` loops are the most common loop type in Python and you will likely use them in nearly every program you write. In a `for` loop, the program repeats the same instructions over a predefined sequence. That is for each element, the set of instructions are carried out, before returning to the top with the next element. It's easiest to explain with an example, the one below prints each element of a list.

```python
my_list = ['hello','world','good','morning']

for word in my_list:
    print(word)
```

This example shows the syntax when starting a `for` loop - `for x in y:` where `y` is some object you can iterate over and `x` is a placeholder name for each element of `y` in turn. In practice the object to iterate over is usually a list, tuple or `range(n)` (which covers the integers 0,...,n-1) but can also work over strings, dictionaries and many other Python objects.

>It's vital that the body of the for loop be indented. Python uses indentation to track the level of different parts of the program. Luckily Jupyter manages indentation automatically whenever a for loop is created, usually by adding 4 spaces or a tab (these should not be mixed within a single idented block!). When you have finished adding to the body of the loop, you should return to the previous indentation level.

`for` loops can be created within other loops like in the example below.

```python
my_list = ['hello','world',' ']
for i in range(2):
    for word in my_list:
        for letter in word:
            print(letter)
```

This example helps to illustrate the order in which the loops proceed. Here we are looping two times (`for i in range(2):`) over all strings in my_list (`for word in my_list:`) and then printing each letter in the string (`for letter in word:`). Another way to think about this is repeating the `print` command many times but where the parameters change:

    i=0, word="hello", letter="h"
    i=0, word="hello", letter="e"
    ...
    i=1, word="world", letter="l"
    i=1, word="world", letter="d"

### `while` loops

The counterpoint to `for` loops are `while` loops. Constructed in a similar way to `for` loops, `while` loops are useful if you do not know how many steps are required ahead of time. The other advantage of a `while` loop is it is comparative in nature, that is whether the loop continues depends on a certain condition being satisfied. Again an example will make this clear.

```python
count = 0
while count < 10:
    count += 1
print(count)
```

This example is checking whether the count has reached 10 (`while count < 10:`), and if it hasn't increments it by 1 (`count += 1`). It is crucial that you change the comparison at some point in the `while` loop, otherwise your loop will run forever! This is one of the most common mistakes with `while` loops, so make sure an update step is the first line you add to the body of your loop.

In many situations both `for` and `while` loops can be used. Due to the possibility of infinite loops and the potentially unknown execution time, `for` loops should be preferred in most circumstances.

## Flow control

Flow control statements are used in programming to automatically include or exclude certain commands whenever certain conditions are met. This section will introduce the basic building blocks which can be assembled to construct programs with as many paths as required.

### `if` statements

`if` statements allow you to check whether a certain condition holds, and only then take action. `if` statements are often useful for checking that you have the expected data type before correcting it, as shown in the example below:

```python
my_num = 3.0
print(type(my_num))
if type(my_num) != int:
    print("Changing types.")
    my_num = int(my_num)
print(type(my_num))
```

```python
my_num = 3
print(type(my_num))
if type(my_num) != int:
    print("Changing types.")
    my_num = int(my_num)
print(type(my_num))
```

As with the `for` and `while` loops before, `if` statements require indentation before their body code is entered. Unlike loops, anything inside this statement is run a single time and only if the specified condition is met.

### `else` statements

`else` statements form a pair with `if` statements, and are used whenever there are two possible cases and each needs to be handled differently. Note that as we saw above, an `if` statement does not require an `else` statement to function. In the example below we split a list of numbers into positive and negative using and `if`-`else` pair.

```python
my_nums = [1,2,3,-1,-2,-3,4,5,6]
pos = []
neg = []

for num in my_nums:
    if num > 0:
        pos += [num]
    else:
        neg += [num]
print(pos)
print(neg)
```

### `elif` statements

With a single `if` statement, you are only allowed a single `else` statement. If there are other mutually-exclusive cases to consider, then you can make use of an `elif` statement. The example below separates a list of strings by whether they start with a vowel, consonant or a number.

```python
vowels = 'aeiou'
nums = '0123456789'

my_list = ['hello','everybody','1stplace','python','coding','python3.x']
v = []
c = []
n = []

for word in my_list:
    if word[0] in vowels:
        v += [word]
    elif word[0] in nums:
        n += [word]
    else:
        c += [word]
print(v)
print(c)
print(n)
```

This example could be extended even further by adding more `elif` statements between the `if` and `else`. This example also shows that the indexing available to lists extends to strings and also introduces the `in` keyword. `in` tests whether a given item is in a sequence. Here we were testing whether the first character was in strings of vowels and numbers we specified, but this can also be used for other data types. Note that the `elif` statements are evaluated in order so you may wish to think carefully about which tests you apply first.

## Functions

At this point you've seen enough to start making your own Python programs for many different purposes. Next we're going to look at functions, which let you call a series of commands in a single statement using prewritten code. We've already seen a couple of examples: `type()`, which returns the datatype of an object; `print()`, which outputs the given string to the display; and `int()` which converts other number types to integers. There are many other in-built functions for different applications, such as `sort()`, which sorts a list or other sequence type. Rather than give an exhaustive list here, I recommend you search the web to find out whether one exists for specific purposes.

In general, functions in Python take a series of arguments (between the brackets) and return one or more data objects. To start a function, you need to use the `def` keyword, an indented block of body code and if you want it to provide an output you must use the `return` keyword. The example below creates a function to merge two lists by alternating the elements. It makes use of the `len()` function, which returns the number of elements in a sequence.

```python
my_list1 = ["a","b","c"]
my_list2 = [1,2,3]

def interleave(list1, list2):
    out_list = []
    for i in range(len(list1)):
        out_list.append(list1[i])
        out_list.append(list2[i])
    return out_list

print(interleave(my_list1,my_list2))
out_list
```

This example demonstrates a few key things to bear in mind when working with functions.
* You can call any of the functions you write in the same way as a default function by passing it the correct number of arguments.
* The inside of a function is a different scope to the rest of your program. This means that variables defined within a function but not returned cannot be accessed outside the function (think of different scopes as separate rooms in a building). Be very careful where you define your variables.

This example also introduces the `list.append()` method. Methods are a special kind of functions that are attached to particular object types. In this case, `append()` is a special function within the `list` class. Methods are usually called using the `object.method()` notation, which should be read as the function `method()` acting on `object` since nearly all methods pass the object itself as their first argument.

Functions do not need to have a `return` statement if you do not want to return any value. In this case if you store the value of the function call you will get a `NoneType` object, a specific means of recording no value. If you find an error related to a `NoneType` object, this is often the cause.

On a related note, methods need to be used quite carefully. There is no standard for the behaviour of methods to follow. In the case above, `list.append()` acts in place and updates the list without any assignment in your code. The `str.lower()` method, which converts an string to all lower case letters, returns the modified string but does not affect the original. The general advice here is whenever you apply a new function or method, look at the documentation to ensure you are using it correctly.

When defining functions it is a good idea to think about future reuse. Two good ways to do this are through comments or docstrings. In Python any code following the `#` character is ignored, so you can use this to annotate and explain your code. You can do this anywhere in Python, not just in functions.

Docstrings on the other hand allow you to add descriptions to functions. In their most common form, three quotes (`"""`) are added around a sentence under the definition explaining the purpose of the function. It can be useful to note the expected input and output types here too. If you ever want to access this string in the future, just type `help(my_func)`. This works on built-in functions too. Here we have added comments and a docstring to the previous example.

```python
def interleave(list1, list2):
    """ Function to merge two lists, alternating their elements. Returns a list. """
    out_list = []  # Use this to build the output list.
    for i in range(len(list1)):  # Iterate over the index of list elements.
        out_list.append(list1[i])  # Add the next element from list1.
        out_list.append(list2[i])  # Add the next element from list2.
    return out_list  # Return the output list.

help(interleave)
```

You can always overwrite the default later by passing a new value for the argument, but if a default value is present you do not need to supply a value for the argument. Let's update the example again the have a default value for `list2`.

```python
def interleave(list1, list2 = ['x','y','z']):
    """ Function to merge two lists, alternating their elements. Returns a list. """
    out_list = []  # Use this to build the output list.
    for i in range(len(list1)):  # Iterate over the index of list elements.
        out_list.append(list1[i])  # Add the next element from list1.
        out_list.append(list2[i])  # Add the next element from list2.
    return out_list  # Return the output list.

print(interleave(my_list1))
print(interleave(my_list1,my_list2))
```

## Reading and writing to file

Everything we have seen so far uses tools entirely within Python, and generates the data we are working with when we need it. But often our data has already been collected by another process or program and we need to import it before we can work on it. Fortunately, Python includes tools to work with many different file types.

At its simplest, all that is required is using the `open()` function to create a file object.

```python
f = open('tale-of-two-cities.txt')
```

The above cell has created a file object of the file `tale-of-two-cities.txt`. It's important to note that this file is located in the same directory as this workbook - Python can only find files if you tell it where to look. You can specify other folders either relative to your current working directory (e.g. subfolders) or by giving the full path (e.g. `C://...`). We're going to make use of the `help` function again to understand what the file object looks like.

```python
help(f)
```

Looking through all this documentation, the `read` function stands out. It allows us to look at a series of characters from the start of the file. Trying this for 10 characters at a time we see the following.

```python
f.read(10)
```

```python
f.read(10)
```

But these calls are returning different characters at the start of the file! The `TextIOWrapper` object created by the `open` function works different to all of the objects you have seen before. In this case, iterating over the first 10 characters with the `read` function has effectively removed them from the file and it now starts at the 11th character for the second call. Because of this behaviour, it is recommended that the `read` function only be used as a check to ensure the file has opened correctly.

There is one very important step when working with files - ensuring that they are closed when you are finished with them. If you leave many files open you will quickly run into errors as Python tries to manage too many objects in memory at the same time. All you need to do when you have finished working with a file is call its `close` method.

```python
f.close()
```

```python
f.read(10)
```

As you can see, trying to read from a closed file produces an error as the content is no longer accessible to Python.

One slightly cleaner way to work with files uses the `with` statement. `with` statements allow you to define and open a file in a single line. The with statement creates a limited scope, like a `for` loop or an `if` statement in which you can work with the file, and most importantly the file is automatically closed when you leave the scope.

```python
with open('tale-of-two-cities.txt') as f:
    print(f.readline())

f.readline()
```

This example shows off two extra tools that are useful when working with files. The first is the `as` statement. `as` creates an alias for a variable, much like assigning it a name. Here it is used to give the file object a convenient shorthand. The second is the `readline()` method, which unsurprisingly reads the next line from the file. If you want to know more about this you can always refer back to the help above.

You may also notice that the printed character from the `readline` looks messed up - this is due to character encoding. The most common type of encoding when downloading content from the internet (such as in this case) is `utf-8`. You can tell Python to decode this correctly using the `encoding='utf-8'` parameter in the `open` function.

If you work with files regularly you will find it useful to know that files can be iterated over just like lists. In this case, each element will be a line as shown in the example below.

```python
with open('tale-of-two-cities.txt',encoding='utf-8') as f:
    i = 0
    for line in f:
        if i < 5:
            print(line)
        i += 1
```

Everything we have seen so far assumes that you are only interested in reading from files. In order to write to a file, you must first open the file in a suitable way by passing the `'w'` argument to the `open` function. After doing this, you can write any string (or anything easily converted to a string) to the file using the `write()` method. In the example below, we write some simple sentences to a new file.

```python
with open('my_new_file.txt', 'w') as f:
    f.write('This is the first line.\n')
    f.write('This is the second line.\n')
    f.write('This starts the third line.')
    f.write('This ends the third line.\nThis is the last line.')
```

The process of creating a new file is managed automatically by Python - if it can't find the file you tell it to write in it will create it. Note that if the file exists, it will be overwritten when you first write something to it. Here you can also see the `\n` command which is necessary to move onto a new line in the file.

This covers the basics of working with files in Python. For more advanced ideas, you should talk to me for ideas or look for one of the many online tutorials covering it in more depth (e.g. https://realpython.com/read-write-files-python/).

```python
words = []  ## Set up storage variable for data from the file.

with open('tale-of-two-cities.txt', 'r', encoding='utf-8') as f:
    for line in f:
        w = line.strip().split(' ')  ## strip removes leading and trailing whitespace, it is useful for tidying text.
        words += w
print('There are %d words in A Tale of Two Cities.' % len(words))  ## Prints a nice statement of the wordcount.

word_counts = {}
## Now iterate through words to count each individual word. Could be done in the other for loop.
for word in words:
    if word_counts.get(word):
        word_counts[word] += 1
    else:
        word_counts[word] = 1

cent_words = [[k,word_counts[k]] for k in word_counts if word_counts[k] > 100]  ## This uses list comprehension to write a for
##  loop on a single line. We check all keys in the dictionary and make a list of pairs where the word is used more than 100 times

## Write the cent_words to file.
with open('cent_words.txt', 'w', encoding='utf-8') as f:
    for word, count in cent_words:  ## The word, count notation unpacks the small lists.
        f.write(word + ':' + str(count) + '\n')

if word_counts['London'] > word_counts['Paris']:
    print('London is used more than Paris.')
if word_counts['Paris'] > word_counts['London']:
    print('Paris is used more than London.')
```

## Plotting

Whenever you're working with data, visualisation can be an excellent tool for understanding and communicating your findings. Python includes robust support for many different types of plots and figures that can be customised to suit any requirements.

Most visualisations in Python make use of `matplotlib`, a module containing many different plotting tools and included with every installation. Modules are files that contain Python code, usually written and shared by members of the community, to provide additional functionality. This extra code can be used in any Python script, but must be added to your file using the `import` command.

```python
import matplotlib.pyplot as plt
```

Let's take a moment to explain what we've done here. The `import` command looks for a module named `matplotlib.pyplot` and makes all the code within available to us. We've called `.pyplot`, a submodule of `matplotlib` since it contains the plotting functions we will be working with. Finally, we give `matplotlib.pyplot` the alias `plt`. This is a fairly standard step to make your code cleaner, but is not strictly necessary. Assigning this alias makes it easier to use the module as we need to use its name to access any functions within.

Let's make this a little clearer with an example. We are going to plot a simple scatter plot using some dummy data.

```python
x = [1,2,3,4]  # Set up the data we are going to use.
y = [1,4,9,16]

plt.scatter(x,y)  # Call the scatter function in matplotlib.pyplot using x and y.
plt.show()  # Display the finished plot. Without this it won't be visible at all!
```

Here we have done three things:
* First, we have set up the data we wish to plot. Lists of numbers are ideal for this but you may have to play around depending on your requirements.
* Second, we have told matplotlib to call the `scatter` function. This produces a simple scatter plot.
* Finally, we have told matplotlib to display the final figure.

Adding multiple plots to the same axes is as simple as calling multiple plot commands before `show()`.

```python
plt.plot(x,y)  # This adds a curve to the figure.
plt.scatter(x,y)

plt.show()
```

If you want to work with different types of plot, `matplotlib` supports many different variants and include examples in their [online documentation](https://matplotlib.org/tutorials/introductory/sample_plots.html#sphx-glr-tutorials-introductory-sample-plots-py).

It is worth taking some time to customise your figures with labels and axes, which can all be handled using similar `matplotlib` commands. Let's add a title and label the axes. We'll also adjust the axes limits to a custom range rather than use the best guess given by `matplotlib`.

```python
plt.plot(x,y)
plt.scatter(x,y)

plt.xlabel('x axis')
plt.xlim([0,5]) # This changes the displayed range to between 0 and 5 on the x axis.
plt.ylim([0,20])  # This changes the displayed range to between 0 and 20 on the y axis.
plt.ylabel('y axis')
plt.title('Title')

plt.show()
```

Now we're starting to get a complete looking figure, there are a few final changes to make to ensure it is as clear as possible. If you take a look through the documentations for the [plot](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot) and [scatter](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html#matplotlib.pyplot.scatter) functions, you will see that there are lots of other options we could include. We're going to focus on methods to make the points and line clearer and add a legend to the figure.

```python
plt.plot(x,y,c='red',label='curve',linestyle='--')  # We're using the c and label arguments to make the curve red and name it.
plt.scatter(x,y,color='blue',label='scatter points')  # linestyle changes the appearance of the line, in this case to dashed.

plt.xlabel('x axis')
plt.xlim([0,5])
plt.ylim([0,20])  
plt.ylabel('y axis')
plt.title('Title')
plt.legend()

plt.show()
```

The following data represents the frequencies of different colours of cars passing a monitoring point over a ten minute interval. Plot it as a bar chart.

```python
data = [10, 5, 6, 7, 8, 9, 15]
colours = ['silver','green','blue','white','yellow','red','black']

plt.bar(colours,data)
plt.show()
```

Now plot the data as a horizontal bar chart, with the bar colour corresponding to the car colour in the `colours` list. Hint: you may have to make the bar borders a different colour.

```python
plt.barh(colours,data,color=colours,edgecolor='black')
plt.show()
```

Plot the car data as a pie chart. Play around with different argument combinations until you find a plot that you are happy with.

```python
plt.pie(data, labels=colours, colors=colours, shadow=True, explode=[0.1]*len(colours))
plt.show()
```

Bonus: use the `matplotlib` subfigure options to arrange the bar chart and pie chart in a single figure.

```python
fig, axs = plt.subplots(1,2)
plt.subplots_adjust(wspace=1)  # Pad space between the subfigures.
axs[0].barh(colours,data,color=colours,edgecolor='black')
axs[1].pie(data, labels=colours, colors=colours, shadow=True, explode=[0.1]*len(colours))
plt.show()
```
